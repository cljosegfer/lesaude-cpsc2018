```python
reference = pd.read_csv('REFERENCE.csv')
reference
```

```python
columns = ['exam_id', 'AF', 'I-AVB', 'LBBB', 'RBBB', 'PAC', 'PVC', 'STD', 'STE']
```

```python
def label2onehot(exam_id, labels):
    onehot = [np.nan, False, False, False, False, False, False, False, False]
    onehot[0] = exam_id
    for label in labels:
        if pd.isna(label):
            continue
        else:
            if label == 1:
                continue # normal
            onehot[int(label) - 1] = True # skip exam_id and normal
    return onehot
```

```python
# data = np.zeros(shape = (len(reference), len(columns)))
data = []
for idx, row in reference.iterrows():
    # exam_id = row['Recording']
    exam_id = int(row['Recording'][1:])
    labels = row['First_label'], row['Second_label'], row['Third_label']
    data.append(label2onehot(exam_id, labels))
```

```python
metadata = pd.DataFrame(data, columns = columns)
metadata
```

```python
metadata.to_csv('metadata.csv', index = None)
```
